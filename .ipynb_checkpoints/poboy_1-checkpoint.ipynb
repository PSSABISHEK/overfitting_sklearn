```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
np.random.seed(0)
```

```python
N = np.random.normal(0, 1)
```

```python
def generate_data(x):
    return np.sin(2 * np.pi * x) + N + 0.1
```

```python
x_data = []
x_data = np.random.uniform(0, 1, 20)
```

```python
x_data.sort()
```

```python
y_data = []
for x in x_data:
    val = generate_data(x)
    y_data.append(val)
```

```python
plt.figure(figsize=(8, 7))
plt.plot(x_data, y_data)
```

```python
split_len = 10
x_train, y_train = x_data[:10], y_data[:10]
x_test, y_test = x_data[10:], y_data[10:]


X_train = x_train.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
train_error = []
test_error = []
```

```python
def polynomial_deg_0():
    np.random.seed(0)
    polynomial_features = PolynomialFeatures(degree=0)
    x_train_poly = polynomial_features.fit_transform(X_train)

    model0 = LinearRegression()
    model0.fit(x_train_poly, y_train)
    y_pred = model0.predict(x_train_poly)

    rmse = np.sqrt(mean_squared_error(y_train, y_pred))
    train_error.append(rmse)

    plt.scatter(X_train, y_train, s=25)
    plt.plot(X_train, y_pred, color='m')
    plt.title('M=0')
    plt.xlabel('X')
    plt.ylabel('y=sin(2*pi*X)+N+0.1')
    plt.show()
```

```python
def polynomial_deg_1():
    np.random.seed(0)
    polynomial_features= PolynomialFeatures(degree=1)
    x_train_poly = polynomial_features.fit_transform(X_train)

    model1 = LinearRegression()
    model1.fit(x_train_poly, y_train)
    y_pred = model1.predict(x_train_poly)

    rmse = np.sqrt(mean_squared_error(y_train, y_pred))
    train_error.append(rmse)
    plt.scatter(X_train, y_train, s=25)
    plt.plot(X_train, y_pred, color='m')
    plt.title('M=1')
    plt.xlabel('X')
    plt.ylabel('y=sin(2*pi*X)+N+0.1')
    plt.show()
```

```python
def polynomial_deg_3():
    np.random.seed(0)
    polynomial_features= PolynomialFeatures(degree=3)
    x_train_poly = polynomial_features.fit_transform(X_train)

    model3 = LinearRegression()
    model3.fit(x_train_poly, y_train)
    y_pred = model3.predict(x_train_poly)

    rmse = np.sqrt(mean_squared_error(y_train, y_pred))
    train_error.append(rmse)

    plt.scatter(X_train, y_train, s=25)
    plt.plot(X_train, y_pred, color='m')
    plt.title('M=3')
    plt.xlabel('X')
    plt.ylabel('y=sin(2*pi*X)+N+0.1')
    plt.show()
```

```python
def polynomial_deg_9():
    np.random.seed(0)
    polynomial_features= PolynomialFeatures(degree=9)
    x_train_poly = polynomial_features.fit_transform(X_train)

    model9 = LinearRegression()
    model9.fit(x_train_poly, y_train)
    y_pred = model9.predict(x_train_poly)

    rmse = np.sqrt(mean_squared_error(y_train, y_pred))
    train_error.append(rmse)
    
    plt.scatter(X_train, y_train, s=25)
    plt.plot(X_train, y_pred, color='m')
    plt.title('M=9')
    plt.xlabel('X')
    plt.ylabel('y=sin(2*pi*X)+N+0.1')
    plt.show()
```

```python
train_error = []
test_error = []
def train_models():
    polynomial_deg_0()
    polynomial_deg_1()
    polynomial_deg_3()
    polynomial_deg_9()
    print(train_error)

train_models()
```
